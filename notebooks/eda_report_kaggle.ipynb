## 1. Product label distribution — concentration + GT fragmentation

```python
top = STATS["product_top15"][::-1]
fig, ax = plt.subplots(figsize=(9, 6))
ax.barh(range(len(top)), [c for _, c in top], color=ORANGE)
ax.set_yticks(range(len(top))); ax.set_yticklabels([k for k, _ in top])
ax.set_xlabel("count in train"); ax.set_title("Top-15 product labels (train)")
cov = sum(c for _, c in top) / STATS["product_nonempty_total"]
ax.text(0.98, 0.04, f"top-15 cover {cov:.0%} of non-empty labels\n{STATS['product_distinct']} distinct products",
        transform=ax.transAxes, ha="right", color=GREEN, fontsize=11)
save(fig, "01_product_longtail"); plt.show()
```

## 2. Empty-GT rate → the ~0.25 baseline floor

```python
e = STATS["empty"]; vals = [e["product"], e["ocr"], e["both"]]
fig, ax = plt.subplots(figsize=(7, 5))
ax.bar(["product empty", "ocr empty", "both empty"], vals, color=[ORANGE, OLIVE, GREEN])
for i, v in enumerate(vals): ax.text(i, v + .005, f"{v:.1%}", ha="center")
ax.set_ylabel("fraction of train"); ax.set_ylim(0, max(vals) * 1.25)
ax.set_title("Empty ground-truth rate (why all-empty already scores ~0.25)")
save(fig, "02_empty_rate"); plt.show()
```

## 3. OCR text length distribution (truncation at 500)

```python
h = STATS["ocr_len"]; edges = np.array(h["edges"]); cnt = np.array(h["counts"])
fig, ax = plt.subplots(figsize=(8, 5))
ax.bar(edges[:-1], cnt, width=np.diff(edges), align="edge", color=OLIVE, edgecolor=GREEN)
ax.axvline(500, color=ORANGE, lw=2, ls="--", label="500-char GT truncation")
ax.set_xlabel("GT ocr_text length (chars)"); ax.set_ylabel("images")
ax.set_title(f"OCR text length — mean {h['mean']:.0f}, median {h['median']:.0f}")
ax.legend(); save(fig, "03_ocr_length"); plt.show()
```

## 4. Diacritic prevalence in GT

```python
h = STATS["diac"]; edges = np.array(h["edges"]); cnt = np.array(h["counts"])
fig, ax = plt.subplots(figsize=(8, 5))
ax.bar(edges[:-1], cnt, width=np.diff(edges), align="edge", color=PEACH, edgecolor=ORANGE)
ax.axvline(h["mean"], color=GREEN, lw=2, label=f"mean {h['mean']:.2f}")
ax.set_xlabel("diacritic marks per letter (GT)"); ax.set_ylabel("images")
ax.set_title("Vietnamese diacritic density in ground-truth text")
ax.legend(); save(fig, "04_diacritic_density"); plt.show()
```

## 5. OCR engine bake-off — CER & diacritic preservation

```python
eng = STATS["engines"]; names = [e[0] for e in eng]
cers = [e[1] for e in eng]; dia = [e[2] for e in eng]
fig, axes = plt.subplots(1, 2, figsize=(13, 5))
axes[0].bar(names, cers, color=[ORANGE, OLIVE, GREEN])
for i, v in enumerate(cers): axes[0].text(i, v + .005, f"{v:.3f}", ha="center")
axes[0].set_ylabel("mean CER (lower=better)"); axes[0].set_title("CER by OCR engine")
axes[1].bar(names, dia, color=[ORANGE, OLIVE, GREEN])
for i, v in enumerate(dia): axes[1].text(i, v + .005, f"{v:.2f}", ha="center")
axes[1].set_ylabel("diacritic marks / letter")
axes[1].set_title(f"Diacritic preservation (GT ref ~{STATS['gt_diac_ref']:.2f})")
for a in axes: a.tick_params(axis="x", rotation=15)
save(fig, "05_engine_bakeoff"); plt.show()
```

## 6. CER bands — the ~27% region-mismatch failure band

```python
bands = ["[0,0.1)", "[0.1,0.3)", "[0.3,0.6)", "[0.6,1.0]"]; fr = STATS["cer_bands"]
fig, ax = plt.subplots(figsize=(8, 5))
ax.bar(bands, fr, color=[GREEN, OLIVE, PEACH, ORANGE])
for i, v in enumerate(fr): ax.text(i, v + .005, f"{v:.0%}", ha="center")
ax.set_ylabel("fraction of text-bearing images")
ax.set_title("CER distribution — the ~27% total-failure band (region mismatch)")
save(fig, "06_cer_bands"); plt.show()
```

## 7. Oracle vs real product F1 — the OCR-noise penalty

```python
fo, fr = STATS["f1_oracle"], STATS["f1_real"]
fig, ax = plt.subplots(figsize=(7, 5))
ax.bar(["oracle (GT text)", "real (VietOCR-FT)"], [fo, fr], color=[GREEN, ORANGE])
ax.annotate("", xy=(1, fr), xytext=(1, fo), arrowprops=dict(arrowstyle="<->", color="black"))
ax.text(1.06, (fo + fr) / 2, f"OCR-noise\npenalty\n{fo - fr:.3f}", va="center", color=ORANGE)
for i, v in enumerate([fo, fr]): ax.text(i, v + .008, f"{v:.3f}", ha="center")
ax.set_ylabel("product token-F1"); ax.set_ylim(0, max(fo, fr) * 1.2)
ax.set_title("Product head: oracle vs real-OCR F1 (in-sample)")
save(fig, "07_oracle_vs_real"); plt.show()
```

## 8. Score progression (public leaderboard)

```python
prog = STATS["progression"]; names = [p[0] for p in prog]; vals = [p[1] for p in prog]
fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(range(len(vals)), vals, "-o", color=ORANGE, lw=2.5, ms=9)
for i, v in enumerate(vals): ax.text(i, v + .003, f"{v:.4f}", ha="center", color=GREEN)
ax.axhline(STATS["rival"], color=OLIVE, ls="--", lw=1.5, label=f"{STATS['rival']} rival (reference)")
ax.set_xticks(range(len(names))); ax.set_xticklabels(names)
ax.set_ylabel("public LB score"); ax.set_title("Public-leaderboard progression")
ax.legend(); save(fig, "08_score_progression"); plt.show()
```

## 9. The experiment log — every idea, judged by one fixed CV

```python
ex = sorted(STATS["experiments"], key=lambda x: x[1])
names = [e[0] for e in ex]; dels = [e[1] for e in ex]
cols = [OLIVE if d >= 0 else ORANGE for d in dels]
fig, ax = plt.subplots(figsize=(9, 6))
ax.barh(range(len(names)), dels, color=cols)
ax.set_yticks(range(len(names))); ax.set_yticklabels(names); ax.axvline(0, color="black", lw=1)
for i, d in enumerate(dels):
    ax.text(d + (0.001 if d >= 0 else -0.001), i, f"{d:+.4f}", va="center",
            ha="left" if d >= 0 else "right", fontsize=10)
ax.set_xlabel("CV composite delta vs baseline")
ax.set_title("Disciplined iteration: one change, one CV verdict")
save(fig, "09_experiment_log"); plt.show()
```

## 10. Phase-1 vs Phase-2 distribution shift

```python
ps = STATS["phase_shift"]; ks = [p[0] for p in ps]; vs = [p[1] for p in ps]
fig, ax = plt.subplots(figsize=(7, 5))
ax.bar(ks, vs, color=[ORANGE, GREEN])
for i, v in enumerate(vs): ax.text(i, v + .005, f"{v:.0%}", ha="center")
ax.set_ylabel("OCR rows with a dominant-family marker")
ax.set_title("Distribution shift: Phase 2 is more diverse")
save(fig, "10_phase_shift"); plt.show()
print("\nAll figures saved to", FIG)
```
